5. (9б). Створіть клас 3D-фігура без атрибутів, з методом get_volume для отримання об'єму фігури, що повертає 0 та __add __, який приймає self та other в якості аргументів, а повертає суму об'ємів 3d-фігур self та other.

```python
class Figure3D:
  def get_volume(self):
    return 0

  def __add__(self,other):
    return self.get_volume() + other.get_volume()
```

6. (11б). Створіть 2 дочірніх класи від 3d-фігура: Циліндр та Куб, які мають атрибути, необхідні для розрахунку об'єму. Визначте метод get_volume в кожному з них так, щоби вони розраховували об'єм у залежності від формули для кожного типу фігури. Створіть обʼєкт класу Циліндр та обʼєкт класу Куб. Виконайте операцію суми за допомогою оператора + між ними, аби знайти сумарний об'єм.

```python
import math

class Cylinder(Figure3D):
  def __init__(self,radius, height):
    self.radius=radius
    self.height=height

  def get_volume(self):
    return math.pi * (self.radius ** 2) * self.height

class Cube(Figure3D):
  def __init__(self, height):
    self.height=height

  def get_volume(self):
    return self.height ** 3

my_cylinder=Cylinder(10,15)
my_cube=Cube(10)

res=my_cylinder+my_cube

print(f"Об'єм куба: {my_cube.get_volume()}")
print(f"Об'єм циліндра: {round(my_cylinder.get_volume(), 2)}")
print(f"Сумарний об'єм: {round(res, 2)}")
```

8. (13б). Створіть клас EWallet з приватними атрибутами balance та id. Реалізуйте методи поповнення та зняття коштів, забезпечивши належну інкапсуляцію. Підказка: використовуйте декоратори property та setter.

```python
class EWallet:
  def __init__(self, id, balance=0.0):
    self.__balance=float(balance)
    self.__id=id

  @property
  def balance(self):
    return self.__balance

  @property
  def id(self):
    return self.__id

  @balance.setter
  def balance(self, value):
    if value < 0:
      print("Баланс не може бути від'ємним!")
    else:
      self.__balance = float(value)

  def deposit(self, amount):
    if amount>0 :
      self.balance +=amount
      print(f"Гаманець поповнено на {amount}. Поточний баланс: {self.balance}.")
    else:
      print("Сума поповнення не може бути від'ємною!")

  def withdraw(self,amount):
    if self.balance>=amount:
      self.balance-= amount
      print(f"Знято з гаманця {amount}. Поточний баланс: {self.balance}.")
    else:
      print("Не достатньо коштів!")

wallet = EWallet("1", 100)

wallet.deposit(50)
wallet.withdraw(30)
wallet.withdraw(200)

wallet.balance = -10
```

9. (11б). Створіть клас ShoppingCart, який містить список об'єктів типу Item. Реалізуйте методи для додавання айтемів, їх видалення та відображення списку айтемів.

```python
class Item:
  def __init__(self, name, price=0.0):
    self.name=name
    self.price=float(price)

  def __repr__(self):
        return f"{self.name} ({self.price} грн)"

class ShoppingCard:
  def __init__(self):
    self.items_list=[]

  def add_item(self, item):
    self.items_list.append(item)
    print(f"Додано: {item.name}")

  def delete_item(self,item_name):
    for item in self.items_list:
      if item.name==item_name:
        self.items_list.remove(item)
        print(f"Видалено: {item_name}")
        return
    print(f"Товар '{item_name}' не знайдено в кошику.")

  def display_items(self):
    if not self.items_list:
      print("Кошик порожній.")
    else:
      print("Товари в кошику:")
      for item in self.items_list:
        print(f"- {item}")

cart = ShoppingCard()

cart.add_item(Item("Кава", 80))
cart.add_item(Item("Сир", 120))
cart.display_items()

cart.delete_item("Кава")
cart.display_items()
```

13. (11б). Створіть клас, що має визначений Iterator Protocol та при кожному виклику next повертає назву дня тижня, поки вони не вичерпаються. Зверніть увагу, що порядок має значення.

```python
class Days:
  def __init__(self):
    self.days=["Понеділок", "Вівторок", "Середа","Четвер", "П'ятниця", "Субота", "Неділя"]
    self.count=0
  def __iter__(self):
        return self
  def __next__(self):
    if self.count < len(self.days):
      day=self.days[self.count]
      self.count+=1
      return day
    else:
      raise StopIteration

week = Days()

for day in week:
    print(day)
```
